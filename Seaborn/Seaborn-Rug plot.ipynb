```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
mart=mart[["Gender","Unit price","Payment","Quantity","Total","gross income"]]
mart.head()
```

```python
basic rug plot   -[ display the mark on axis ]
```

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x="Unit price")
```

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x="Unit price",y="gross income")
```

```python
# group it by a categorical variable using hue
```

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x="Unit price",y="gross income",hue="Gender")
```

```python
# change height 
```

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart,x="Unit price",y="gross income",hue="Gender",height=0.05)
```

```python
#combine with a KDE plot
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart,x="Unit price",y="gross income",hue="Gender",fill=True)
sns.rugplot(data=mart,x="Unit price",y="gross income",hue="Gender")
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=mart,x="Unit price",y="gross income",hue="Gender")
sns.rugplot(data=mart,x="Unit price",y="gross income",hue="Gender")
```

```python
# show rugs outside of the axis
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart,x="Unit price",y="gross income",hue="Gender",fill=True)
sns.rugplot(data=mart,x="Unit price",y="gross income",hue="Gender",height=-0.01,clip_on=False)
```

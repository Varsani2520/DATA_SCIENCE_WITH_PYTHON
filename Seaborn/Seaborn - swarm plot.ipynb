```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
# create basic swarmplot on the total column
```

```python
sns.swarmplot(data=mart,y="Total")
```

```python
# create a swarmplot group by categories
```

```python
sns.swarmplot(x="Payment",y="Total",data=mart)
```

```python
sns.swarmplot(x="Payment",y="Total",data=mart,hue="Gender")
```

```python
# showing hues separetly on the categorical axis
```

```python
sns.swarmplot(x="Payment",y="Total",data=mart,hue="Gender",dodge=True)
```

```python
# styling a swarm - change the marker,size, color,edge color..
```

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x="Payment",y="Total",data=mart,marker="*",size=10,color="green",edgecolor="black")
```

```python
# overlay a swarm plot on box plot
```

```python
sns.swarmplot(x="Payment",y="Total",data=mart)
sns.boxplot(x="Payment",y="Total",data=mart)
```

```python
# overlay warm on violin plot
```

```python
sns.swarmplot(x="Payment",y="Total",data=mart)
sns.violinplot(x="Payment",y="Total",data=mart)
```

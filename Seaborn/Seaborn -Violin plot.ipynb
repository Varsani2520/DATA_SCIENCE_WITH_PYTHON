```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
# create a basic violinplot
```

```python
sns.violinplot(y="Total",data=mart)
```

```python
# create a violin plot on two categorical and one numerical variable and use split
```

```python
sns.violinplot(x="Payment",y="Total",data=mart,hue="Gender")
```

```python
sns.violinplot(x="Payment",y="Total",data=mart,hue="Gender",split=True)
```

```python
#change the box in the violin plot to horizontal lines
```

```python
sns.violinplot(x="Payment",y="Total",data=mart,hue="Gender",split=True,inner="quart")
```

```python
sns.violinplot(x="Payment",y="Total",data=mart,hue="Gender",split=True,inner="stick")
```

```python
# change the amount of smoothing using bw attribute
```

```python
sns.violinplot(x="Payment",y="Total",data=mart,hue="Gender",split=True,inner="quart",bw=1)
```

```python
# cut out the extreme value
```

```python
sns.violinplot(x="Payment",y="Total",data=mart,hue="Gender",split=True,inner="quart",cut=0)
```

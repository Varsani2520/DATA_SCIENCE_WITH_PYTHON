```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(data=iris)
```

```python
# change diagonal plot to KDE ,Hist or none
```

```python
sns.pairplot(data=iris,diag_kind="kde")
```

```python
# changing the non diagonal plot KIND to scatter, kde ,hist or reg
```

```python
sns.pairplot(data=iris,kind="reg")
```

```python
# add hue
```

```python
sns.pairplot(data=iris,kind="reg",hue="species")
```

```python
# pair plot for specific list of var ( diag_kind=None)
```

```python
sns.pairplot(data=iris,x_vars=["sepal_width","petal_width","petal_length"],y_vars="sepal_length")
```

```python
# showing only the  lower triangle usign corner arg
```

```python
sns.pairplot(data=iris,corner=True) # cut upr corner
```

```python
sns.pairplot(data=iris,corner=False)
```

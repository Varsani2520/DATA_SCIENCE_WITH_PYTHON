```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
# create a strip plot
```

```python
sns.stripplot(data=mart,y="Total")
```

```python
# expand markers in strip to plot using jliter
```

```python
sns.stripplot(data=mart,y="Total",jitter=0.2)
```

```python
# draw a line around the points using linewidth
```

```python
sns.stripplot(data=mart,y="Total",linewidth=0.5)
```

```python
#include third categorical variable with hue
```

```python
sns.stripplot(data=mart,y="Total",x="Payment",jitter=0.2,hue="Gender",linewidth=0.3)
```

```python
# separate each level of the hue using dodge
```

```python
sns.stripplot(data=mart,y="Total",x="Payment",jitter=0.2,hue="Gender",linewidth=0.3,dodge=True)
```

```python
# draw a strip on top of violin plot
```

```python
sns.stripplot(data=mart, y="Total", x="Payment")
sns.violinplot(data=mart, y="Total", x="Payment",color='gray')
```

```python
# draw a strips on top of box plot
```

```python
sns.stripplot(data=mart, y="Total", x="Payment",color="black")
sns.boxplot(data=mart, y="Total", x="Payment",color='gray')
```

```python
# import the required libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the data
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
# create a basic countplot to get number of transcation for each of the product line and change the figure size 
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Product line",data=mart)
```

```python
# make a horizontal bar plot
```

```python
sns.countplot(y="Product line",data=mart)
```

```python
# add hue to get the count on two categories i.e product line and gender
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Product line",hue="Gender",data=mart)
```

```python
# using different color palette
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Product line",hue="Gender",data=mart,palette="gist_gray")
```

```python
# change style using facecolor,linewidth,and edge color
```

```python
plt.figure(figsize=(15,5))

sns.countplot(
    x="Product line",
    hue="Gender",
    data=mart,
    palette="pastel",  # Choose a predefined color palette
    edgecolor="black",  # Set edge color for bars
    width=0.8  # Standard bar width
)

plt.title("Distribution of Product Line by Gender")
plt.xlabel("Product Line")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

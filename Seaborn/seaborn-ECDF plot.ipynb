```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
mart=mart[["Gender","Unit price","Payment","Quantity","Total","gross income"]]
mart.head()
```

```python
# sns.ecdfplot(data=mart,x="gross income",hue="Gender")
sns.ecdfplot(data=mart,x="gross income",hue="Gender",stat="count")
```

```python
sns.ecdfplot(data=mart,y="gross income",hue="Gender",stat="count")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
mart=mart[["Gender","Unit price","Payment","Quantity","Total","gross income"]]
mart.head()
```

```python
Â£ create basic KDE plot for total cpolumn
```

```python
sns.kdeplot(data=mart,x="Total")
```

```python
 create kde plot for all the numeric variables in dataframe
```

```python
sns.kdeplot(data=mart)
```

```python
adjust the smoothing using bw_adjust9band width adjust)
```

```python
sns.kdeplot(data=mart,x="Total",bw_adjust=0.2)   ## The 'bw_adjust' parameter adjusts the smoothness of the KDE curve.
```

```python
group the KDE on a acategorical variable
```

```python
sns.kdeplot(data=mart,x="Total",hue="Payment")
```

```python
stack kde on a category using MULTIPLE argument
```

```python
sns.kdeplot(data=mart,x="Total",hue="Payment",multiple="stack")
```

```python
sns.kdeplot(data=mart,x="Total",hue="Payment",log_scale=True)  #The x-axis or y-axis (or both) will use a logarithmic scale instead of a linear scale.
```

```python
changing style of hued KDE using linewidth,palette ,alpha etc...
```

```python
sns.kdeplot(data=mart,x="Total",hue="Payment",multiple="stack",linewidth=5,palette="Dark2",alpha=0.2)
```

```python
create abivariable KDE
```

```python
sns.kdeplot(data=mart,x="Unit price",y="gross income")
```

```python
group the bivariable KDE  on a categorical variable and show the contours
```

```python
# sns.kdeplot(data=mart,x="Unit price",y="gross income",hue="Gender",fill=True)
sns.kdeplot(data=mart,x="Unit price",y="gross income",hue="Gender",levels=10,thresh=0.5) 
#  levels : Controls how many contour lines (or specific density levels) are drawn.
#This will hide areas with density below 20%.
```

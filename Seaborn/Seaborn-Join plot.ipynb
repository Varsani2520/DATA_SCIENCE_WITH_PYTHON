```python
It helps visualize both individual distributions and the relationship between two variables in one plot.
    
It is useful for exploratory data analysis (EDA) to understand patterns, correlations, and data distributions.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length")

# kde marginal distribution plot x and y, relationsal plot
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",kind="kde")

# kde marginal distribution plot x and y, relationsal plot
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",kind="hist")

# kde marginal distribution plot x and y, relationsal plot
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",kind="reg")

# kde marginal distribution plot x and y, relationsal plot
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",kind="resid")

# kde marginal distribution plot x and y, relationsal plot
```

```python
# grouping based on a categorical variable
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",hue="species")
```

```python
# formating all and indivisual plots
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",color="red",
             hue="species",palette="Dark2")
```

```python
sns.jointplot(data=iris,x="sepal_length",y="petal_length",joint_kws=dict(marker="*",color="red"),
             marginal_kws=dict(color="green",kde=True))
```

```python
# adjust height , ratio,space and show / hide the marginal ticks
```

```python
# sns.jointplot(data=iris,x="sepal_length",y="petal_length",height=3)
sns.jointplot(data=iris,x="sepal_length",y="petal_length",ratio=3,space=5,marginal_ticks=True)
```

```python
# plot kde  adn rug on top of joint plot
```

```python
l=sns.jointplot(data=iris,x="sepal_length",y="petal_length",color="red")
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot)
```

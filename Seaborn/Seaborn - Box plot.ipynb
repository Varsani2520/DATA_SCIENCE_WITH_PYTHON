```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
# create a basic BOX plot on one NUMERIC variable
```

```python
sns.boxplot(x="Total",data=mart)
sns.boxplot(y="Total",data=mart,width=0.2)
```

```python
# create a basic box plot on one one numeric variable by a categorical variable
```

```python
sns.boxplot(y="Total",x="Payment",data=mart)
```

```python
# create box ploot on one numerical by two categorical variable using HUE attribute
```

```python
sns.boxplot(y="Total",x="Payment",data=mart,hue="Gender")
```

```python
# add mean marker in box plot using showmeans attribute and change its style using meanprops
```

```python
# sns.boxplot(y="Total",x="Payment",data=mart,hue="Gender",showmeans=True)
sns.boxplot(y="Total",x="Payment",data=mart,hue="Gender",showmeans=True,meanprops={"marker":"o","markerfacecolor":"white","markersize":"5","markeredgecolor":"black"})
```

```python
# make horizontal box plot
```

```python
# sns.boxplot(y="Total",x="Payment",data=mart,hue="Gender",showmeans=True)
sns.boxplot(x="Total",y="Payment",data=mart,hue="Gender",showmeans=True,meanprops={"marker":"o","markerfacecolor":"white","markersize":"5","markeredgecolor":"black"})
```

```python
# change the color palette
```

```python
# sns.boxplot(y="Total",x="Payment",data=mart,hue="Gender",showmeans=True)
sns.boxplot(y="Total",x="Payment",data=mart,hue="Gender",palette="CMRmap",linewidth=5,showmeans=True,meanprops={"marker":"o","markerfacecolor":"white","markersize":"5","markeredgecolor":"black"})
```

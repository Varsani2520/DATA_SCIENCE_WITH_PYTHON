```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
mart=mart[["Gender","Unit price","Payment","Quantity","Total","gross income","Branch"]]
mart.head()
```

```python
sns.displot(data=mart,x="Total")
```

```python
sns.displot(data=mart,x="Total",kde=True)
```

```python
sns.displot(data=mart,x="Total",kde=True,rug=True)
```

```python
# kws= keywords
sns.displot(data=mart,x="Total",kde=True,rug=True,rug_kws={"height":0.05},kde_kws={"bw_adjust":3})    
```

```python
# kws= keywords
sns.displot(data=mart,x="Total",kde=True,rug=True,rug_kws={"height":0.05},kde_kws={"bw_adjust":3},hue="Payment",col="Gender",multiple="stack",row="Branch")    
```

```python
# kws= keywords
sns.displot(data=mart,x="Total",kde=True,rug=True,rug_kws={"height":0.05},kde_kws={"bw_adjust":3},hue="Payment",row="Gender",multiple="stack",col="Branch")    
```

```python
# kws= keywords
sns.displot(data=mart,x="Total",kind="ecdf",hue="Payment",col="Gender",row="Branch")    
```

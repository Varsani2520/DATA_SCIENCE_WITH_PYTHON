```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read dataset
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
What is avg tip for each day of the week ?
```

```python
sns.barplot(x="day",y="tip",data=df,ci=False)
plt.title("Average tip by day")
```

```python
Compare the total bill amount for lunch and dinner.
```

```python
sns.barplot(x="time",y="total_bill",data=df)
plt.title("total bill by time")
```

```python
What is the total tip amount for smokers vs. non-smokers?
```

```python
sns.barplot(x="smoker",y="total_bill",data=df)
plt.title("total tip amount: Smoket vs Non-smoker")
```

```python
How many customers visited on each day of the week?
```

```python
sns.countplot(x="day",data=df)
plt.title("customer visits by day")
```

```python
How many male vs. female customers were there?
```

```python
sns.countplot(x="sex",data=df)
```

```python
Compare the count of smokers vs. non-smokers.
```

```python
sns.countplot(x="smoker",data=df)
```

```python
What is the distribution of total bill for each day of the week?
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
```

```python
How does the tip amount vary between male and female customers?
```

```python
sns.boxplot(x="sex",y="tip",data=df)
```

```python
What is the difference in total bill between lunch and dinner?
```

```python
sns.boxplot(x="time",y="total_bill",data=df)
```

```python
Visualize the distribution of total bill for each day of the week with gender as a hue.
```

```python
sns.violinplot(data=df, x="day", y="total_bill", hue="sex", split=True)
plt.title("Distribution of Total Bill by Day (Gender Split)")
plt.show()
```

```python
Show the distribution of tips for smokers vs. non-smokers.
```

```python
sns.violinplot(data=df, x="tip", y="smoker", split=True)
plt.show()
```

```python
How does the tip percentage (tip/total_bill) vary for different days?
```

```python
Visualize individual total bills for each day.
```

```python
sns.stripplot(data=df, x="day", y="total_bill", jitter=True)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
Plot the tip amount for each gender on different days.
```

```python
sns.stripplot(data=df,x="day",y="tip",hue="sex",jitter=True,dodge=True)
```

```python
Compare the total bill amount for smokers vs. non-smokers.
```

```python
sns.stripplot(data=df, x="smoker", y="total_bill", jitter=True)
plt.title("Total Bill: Smokers vs. Non-Smokers")
plt.show()
```

```python
Show the tip amount for customers based on the size of their group.
```

```python
sns.swarmplot(data=df, x="size", y="tip")
plt.title("Tip Amount by Group Size")
plt.show()
```

```python
Compare the total bill for lunch and dinner using a swarm plot.
```

```python
sns.swarmplot(data=df, x="time", y="total_bill")
plt.title("Total Bill: Lunch vs. Dinner")
plt.show()
```

```python
Visualize the total bill for smokers vs. non-smokers across all days.
```

```python
sns.swarmplot(data=df, x="smoker", y="total_bill", hue="day")
plt.title("Total Bill: Smokers vs. Non-Smokers by Day")
plt.show()
```

```python
What is the average tip amount for each day of the week?
```

```python
sns.pointplot(data=df, x="day", y="tip")
plt.title("Average Tip Amount by Day")
plt.show()
```

```python
Compare the average total bill for lunch and dinner.
```

```python
sns.pointplot(data=df, x="time", y="total_bill")
plt.title("Average Total Bill: Lunch vs. Dinner")
plt.show()
```

```python
What is the trend of average tips for different group sizes?
```

```python
sns.pointplot(data=df, x="size", y="tip")
plt.title("Average Tip Amount by Group Size")
plt.show()
```

```python
How does the tip percentage vary across days for smokers vs. non-smokers?
```

```python
sns.barplot(data=df,x="day",y="tip",hue="smoker",errorbar=None)
```

```python
Compare total bill and tip amounts across genders and time (lunch vs. dinner).
```

```python
sns.catplot(kind="bar",y="total_bill",hue="time",col="day",x="sex",data=df)
```

```python
Analyze total bill distribution for different group sizes using a box plot and strip plot.
```

```python
sns.boxplot(data=df,y="total_bill",x="size",showfliers=False)
sns.stripplot(data=df,y="total_bill",x="size",color="black",jitter=True)
```

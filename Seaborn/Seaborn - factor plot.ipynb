```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
factor plot disprected . so now is catplot

using catplot , you can perform swarm,bar,strip,violin,box plot

default show strip plot
```

```python
sns.catplot(data=mart,x="Payment",y="Total")
```

```python
sns.catplot(data=mart,x="Payment",y="Total",kind="box")
```

```python
sns.catplot(data=mart,x="Payment",y="Total",kind="violin")   
#with cuting edge,increasing width,inner portile
```

```python
# sns.catplot(data=mart,x="Payment",y="Total",hue="Gender",kind="violin")
sns.catplot(data=mart,x="Payment",y="Total",col="Gender",kind="violin",,row="Customer type")
```

```python
sns.catplot(data=mart,x="Payment",y="Total",col="Gender",kind="violin",row="Customer type",color="gray")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
mart=mart[["Gender","Unit price","Payment","Quantity","Total","gross income"]]
mart.head()
```

# create basic histogram on X or Y axis

```python
sns.histplot(data=mart,x="Total")
```

```python
sns.histplot(data=mart,y="Total")
```

```python
# change the bins width using binwidth argument
```

```python
sns.histplot(data=mart,y="Total",binwidth=150)
```

```python
# change number of bins and interval
```

```python
plt.figure(figsize=(15,5))
sns.histplot(data=mart,x="Total",bins=np.arange(0,1100,50))   #0 to 1100 range and 50 skip so 1 bin=50

plt.xticks(np.arange(0,1100,50))
```

```python
# combine with KDE
```

```python
sns.histplot(data=mart,x="Total",kde=True)  # line show the distribution 
```

```python
# use categorical variable in hue and stack it using multiple argument
```

```python
# sns.histplot(data=mart,x="Total",hue="Payment")
sns.histplot(data=mart,x="Total",hue="Payment",multiple="stack")
```

```python
# make a strip /poly plot using element argument and change the fill
```

```python
sns.histplot(data=mart,x="Total",element="poly",fill=False)
# sns.histplot(data=mart,x="Total",element="step",fill=False)
# sns.histplot(data=mart,x="Total",element="poly",fill=False)
# sns.histplot(data=mart,x="Total",element="poly",hue="Payment")
```

```python
use categorical variable and shrink it
```

```python
# sns.histplot(data=mart,x="Payment",shrink=0.7)
sns.histplot(data=mart,x="Payment",stat="probability")
```

```python
# load datset
```

```python
mart=pd.read_csv("https://raw.githubusercontent.com/sersun/supermarket-sales-analysis/refs/heads/main/supermarket_sales.csv")
# mart.shape
# mart.columns
mart.head()
```

```python
# basic bar plot
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart)
```

```python
# add Hue to the barplot
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,hue="Gender")
```

```python
# make a horizonatal barplot
```

```python
plt.figure(figsize=(15,10))
sns.barplot(y="Product line",x="Total",data=mart,hue="Gender")
```

```python
# plot the bars in given order [alphabetically]
```

```python
# x=mart["Product line"].sort_values()
# x.unique()


plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,hue="Gender",order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"])
```

```python
x=mart["Product line"].sort_values()
x.unique()
```

```python
# add cap on error bar
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,hue="Gender",order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"],capsize=0.8)
```

```python
# remove error bar using ci
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,hue="Gender",order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"],errorbar=None)
```

# change the bar colouring using COLOR attribute

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"],color="black")
```

```python
# change the bar colouring using palette attribute
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"],palette="cividis",hue="Gender")
```

```python
# using saturation Parameter
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"],palette="cividis",hue="Gender",saturation=0.5)
```

```python
# change default aggregation method using estimator parameter
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,order=['Electronic accessories', 'Fashion accessories',
       'Food and beverages', 'Health and beauty', 'Home and lifestyle',
       'Sports and travel'],hue_order=["Male","Female"],palette="cividis",hue="Gender",estimator=sum)
```
